```python
import pandas as pd 
import pipewise 
```

```python
df = pd.DataFrame({
    'ac': [1, 2, 3, 4],
    'bc': [5, 6, 7, 8],
    'lc': [[1,2,3], [2,5,1], [7,4,2], [20,34,12]],
    'sc': ['a,b,c', 'b,2,e', 'c,3,5', 'd,2,3']
    })
df    
```

```python
pipe = pipewise.Pipewise(df)


@pipe.register(outputs='sc_list')
def proc_string_as_list(sc):
    return sc.split(',')
    
```

```python
df_result = pipe.run()
```
